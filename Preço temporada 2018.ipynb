```python
import pandas as pd
```

```python
dados = pd.read_csv('transfermarkt_fbref_201819.csv', sep = ';')
```

```python
dados.columns
```

```python
dados = dados [['player', 'nationality','position','squad','age', 'value', 'birth_year' ]]
```

```python
dados.columns
```

```python
renomear = {'player':'jogador', 'nationality':'nacionalidade', 'position':'posicao', 'squad':'time', 'age':'idade', 'value':'valor',
       'birth_year':'ano_nascimento'}
```

```python
liga = dados.rename(columns = renomear)
```

```python
liga.head()
```
